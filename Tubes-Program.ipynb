# Program Iterative

```python
def is_palindrome(s):
    normalized = ''.join(char.lower() for char in s if char.isalnum())
    
    left, right = 0, len(normalized) - 1
    while left < right:
        if normalized[left] != normalized[right]:
            return False
        left += 1
        right -= 1
        
    return True
```

```python
for case in test_cases:
    result = is_palindrome(case)
    print(f"'{case}' -> {'Palindrom' if result else 'Bukan Palindrom'}")
```

# Program Rekursif

```python
def is_palindrome_recursive(s):
    """
    Fungsi rekursif untuk mendeteksi apakah sebuah string adalah palindrom.
    """
    # Normalisasi string: hilangkan spasi, ubah ke huruf kecil, dan abaikan simbol
    normalized_s = ''.join(char.lower() for char in s if char.isalnum())
    
    # Fungsi rekursif untuk membandingkan karakter
    def helper(substring):
        # Basis rekursi: jika panjang string 0 atau 1, maka palindrom
        if len(substring) <= 1:
            return True
        # Bandingkan karakter pertama dan terakhir
        if substring[0] != substring[-1]:
            return False
        # Panggil rekursi pada substring tanpa karakter pertama dan terakhir
        return helper(substring[1:-1])
    
    # Panggil helper dengan string yang sudah dinormalisasi
    return helper(normalized_s)


for case in test_cases:
    result = is_palindrome_recursive(case)
    print(f"'{case}' -> {'Palindrom' if result else 'Bukan Palindrom'}")
```

```python
# Inisialisasi list untuk menyimpan waktu eksekusi
iterative_times = []
recursive_times = []

# Mengukur waktu untuk setiap string dalam test_cases
for test_string in test_cases:
    # Algoritma iteratif
    start_time = time.time()
    is_palindrome_result = is_palindrome(test_string)
    iterative_time = time.time() - start_time
    iterative_times.append(iterative_time)

    # Algoritma rekursif
    start_time = time.time()
    is_palindrome_recursive_result = is_palindrome_recursive(test_string)
    recursive_time = time.time() - start_time
    recursive_times.append(recursive_time)

# Membuat grafik
plt.figure(figsize=(12, 6))

# Membuat garis
x_points = range(len(test_cases))
plt.plot(x_points, iterative_times, 'b-', linewidth=2, label='Algoritma Iteratif')
plt.plot(x_points, recursive_times, 'r-', linewidth=2, label='Algoritma Rekursif')

# Mempercantik grafik
plt.xlabel('Nomor Test Case')
plt.ylabel('Waktu Eksekusi (detik)')
plt.title('Perbandingan Runtime Algoritma Palindrom')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Menambahkan titik pada setiap data
plt.plot(x_points, iterative_times, 'bo', markersize=6)
plt.plot(x_points, recursive_times, 'ro', markersize=6)

# Menampilkan statistik
avg_iterative = np.mean(iterative_times)
avg_recursive = np.mean(recursive_times)
total_iterative = sum(iterative_times)
total_recursive = sum(recursive_times)

stats_text = f'Rata-rata Iteratif: {avg_iterative:.6f}s\n'
stats_text += f'Rata-rata Rekursif: {avg_recursive:.6f}s\n'
stats_text += f'Total Iteratif: {total_iterative:.6f}s\n'
stats_text += f'Total Rekursif: {total_recursive:.6f}s\n'
stats_text += f'Selisih Total: {abs(total_recursive - total_iterative):.6f}s'
plt.figtext(0.02, 0.02, stats_text, fontsize=8, bbox=dict(facecolor='white', alpha=0.8))

# Mengatur batas y agar grafik lebih mudah dibaca
plt.margins(x=0.02)
plt.tight_layout()
plt.show()

# Cetak hasil perbandingan
print("\nHasil Perbandingan Runtime:")
print("-" * 50)
print(f"Total Waktu Iteratif: {total_iterative:.6f} detik")
print(f"Total Waktu Rekursif: {total_recursive:.6f} detik")
print(f"Rata-rata Waktu Iteratif: {avg_iterative:.6f} detik")
print(f"Rata-rata Waktu Rekursif: {avg_recursive:.6f} detik")
print(f"Selisih Total Waktu: {abs(total_recursive - total_iterative):.6f} detik")
print("-" * 50)
print(f"Algoritma {'Iteratif' if total_iterative < total_recursive else 'Rekursif'} lebih cepat")
print(f"Persentase perbedaan: {abs(total_recursive - total_iterative) / min(total_iterative, total_recursive) * 100:.2f}%")

# Menampilkan detail setiap test case
print("\nDetail per Test Case:")
print("-" * 80)
print(f"{'No.':<4} {'Test Case':<30} {'Iteratif (s)':<15} {'Rekursif (s)':<15} {'Palindrom':<10}")
print("-" * 80)
for i, (test, iter_time, rec_time) in enumerate(zip(test_cases, iterative_times, recursive_times)):
    is_pal = "Ya" if is_palindrome(test) else "Tidak"
    print(f"{i+1:<4} {test[:27]+'...' if len(test)>27 else test:<30} {iter_time:<15.6f} {rec_time:<15.6f} {is_pal:<10}")
```
